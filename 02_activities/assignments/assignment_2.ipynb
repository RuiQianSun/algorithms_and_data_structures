My partner's example is a binary tree with 1 as the root, 2 and 2 as the children of 1, 3 and 5 as the children of the left node 2, and 6 and 7 as the children of the right node 2. Since there are four leaves (3, 5, 6, 7) there are four paths from root to leaf.


-   Copy the solution your partner wrote. 


```python
# Your answer here

# Definition for a binary tree node.
from typing import List

class TreeNode(object):
     def __init__(self, val = 0, left = None, right = None):
         self.val = val
         self.left = left
         self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    """ Returns all root-to-leaf paths in a binary tree """
    result = []
   
    def dfs(node, path):
        if not node:
            return
        path.append(node.val) 
       
        if not node.left and not node.right:  # Leaf node reached
            result.append(list(path))
        else:
            dfs(node.left, path)
            dfs(node.right, path)
       
        path.pop()  # Backtrack
   
    dfs(root, [])
    return result

def build_tree_from_bfs(arr: List[int]) -> TreeNode:
    """ Builds a binary tree from BFS input list """
    if not arr:
        return None
   
    nodes = [TreeNode(val) for val in arr]
    for i in range(len(arr)):
        left_index = 2 * i + 1
        right_index = 2 * i + 2
        if left_index < len(arr):
            nodes[i].left = nodes[left_index]
        if right_index < len(arr):
            nodes[i].right = nodes[right_index]
   
    return nodes[0]  # Root node
```


-   Explain why their solution works in your own words.


My partner's solution works because they traverse a binary tree using DFS and store the value of each popped node into a list called "path". If the node has no children, they have found a leaf so they append the current "path" to the "result". The current node is then popped from "path" to backtrack one node and traverse another path to a different leaf. If there are no other children at the current depth the code will backtrack up a level until every path to leaf is appended to "result" which will be a list of lists.


-   Explain the problem’s time and space complexity in your own words.


The problem's time complexity is O(n), because we have to traverse to every leaf and every leaf is connected to the root so we have to check every value in the binary tree.

The problem's space complexity is O(n), because in the worst case we could have to store every value in memory if the height of the tree is n.
